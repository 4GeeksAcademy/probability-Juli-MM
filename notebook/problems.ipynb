## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random
import numpy as np


#print(dado_1)
#print(dado_2)
#print(suma_tiradas)

def n_repeticiones(n_tiradas=1000):
    contador=0
    for i in range (n_tiradas):
        dado_1=random.randint(1,6)
        dado_2=random.randint(1,6)
        suma_tiradas=dado_1+dado_2

        if suma_tiradas %2 ==0 or suma_tiradas>7:
            contador +=1
        print(i,suma_tiradas,contador)
    
    return contador / n_tiradas
n_repeticiones

print(f"La probabilidad de que la suma de las tiradas salga un numero par o un numero mayor que 7 es del {np.round(n_repeticiones(),2) * 100 }%")
  
    
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i >= 10) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
def coger_bolas(n_repeticiones = 1000):
    contador_1 = 0
    contador_2 = 0

    for i in range(n_repeticiones):
        colores = []
        for j in range(5):
            colores.append(ball_box[np.random.randint(0, 59)])
        #print(colores)
            
        # Convert list to Numpy array for better filtering
        colores = np.array(colores)
        
        bolas_blancas = sum(colores == "White")
        bolas_rojas = sum(colores == "Red")
        bolas_verdes = sum(colores == "Green")

        # Decide if we should add it to the count:
        if (bolas_blancas == 3) and (bolas_rojas == 2):
            contador_1 += 1
        
        if (bolas_blancas == 5) or (bolas_rojas == 5) or (bolas_verdes == 5):
            contador_2 += 1
    
    probability_1 = contador_1 / n_repeticiones
    probability_2 = contador_2 / n_repeticiones

    return probability_1, probability_2

coger_bolas

print(f"La probabilidad de que salgan 3 bolas blancas y 2 rojas es del: {np.round(coger_bolas()[0] * 100, 2)}%")
print(f"La probabilidad de que salgan todas las bolas del mismo color es del: {np.round(coger_bolas()[1] * 100, 2)}%")
```
