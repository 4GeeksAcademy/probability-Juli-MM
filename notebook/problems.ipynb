## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import random

def probability (total_experimentos=1000):

    contar_eventos=0

    for _ in range (total_experimentos):

        dado_1 = np.random.randint(1,6)
        #print(dado_1)
        dado_2 = np.random.randint(1,6)
        #print(dado_2)

        sum_dados = dado_1 + dado_2
        #print(sum_dados)

        if (sum_dados > 7) or (sum_dados % 2 == 0):
            contar_eventos +=1
            #print(contar_eventos)

    return (contar_eventos/total_experimentos)*100

probability(total_experimentos=1000)

print("La probabilidad de que pase el evento en cuestión es del", np.round(probability(),2),"%")
  
    
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import numpy as np
import random

caja_bolas={}

#Diccionario Colores:

for i in range(60):

    if i < 10: caja_bolas[i] = "Blanca"
    elif i>=10 and i<30: caja_bolas[i] = "Roja"
    else: caja_bolas[i] = "Verde"

print(caja_bolas)

def probabilidades (eventos = 1000):
    contador_1 = 0
    contador_2 = 0 
    for _ in range (eventos):

#Definición coger 5 bolas:
        
        colores=[]
        for b in range(5):
            colores.append(caja_bolas[np.random.randint(0,59)])
        #print(colores)

        colores = np.array(colores)

#Conteo cantidad bolas mismo color:
        
        bolas_blancas = sum(colores == "Blanca")
        bolas_rojas = sum(colores == "Roja")
        bolas_verdes = sum(colores == "Verde")

#Contador para condición 1:
        
        if (bolas_blancas == 3) and (bolas_rojas == 2):
            contador_1 += 1
        
#Contador para condición 2:
      
        if (bolas_blancas == 5) or (bolas_rojas == 5) or (bolas_verdes == 5):
            contador_2 += 1

#Calculo probabilidades:
            
        probability_1 = (contador_1 / eventos) * 100
        probability_2 = (contador_2 / eventos) * 100 
        #print(probability_1)
        #print(probability_2)

    return probability_1 , probability_2

probabilidades()

print(f"La probabilidad de que salgan 3 bolas blancas y 2 rojas es del, {np.round(probabilidades()[0],2)}% \nLa probabilidad de que salgan todas las bolas del mismo color es, {np.round(probabilidades()[1],2)}%")
```
